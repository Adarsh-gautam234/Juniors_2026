```python
import numpy as np 
import pandas as pd 
```

```python
def load_data():
    train = pd.read_csv("train.csv")
    test = pd.read_csv("test.csv")
    
    X = train.iloc[:, :-1].values
    y = train.iloc[:, -1].values.reshape(-1, 1)
    
    return X, y, test.values
```

```python
def ols_with_intercept(X, y):
    ones = np.ones((X.shape[0], 1))
    X_new = np.hstack((ones, X))
    
    w = np.linalg.inv(X_new.T @ X_new) @ X_new.T @ y
    return w
```

```python
def ols_no_intercept(X, y):
    w = np.linalg.inv(X.T @ X) @ X.T @ y
    return w
```

```python
def predict_with_intercept(X, w):
    ones = np.ones((X.shape[0], 1))
    X_new = np.hstack((ones, X))
    return X_new @ w
```

```python
def predict_no_intercept(X, w):
    return X @ w
```

```python
def compute_metrics(y_true, y_pred):
    mse = np.mean((y_true - y_pred)**2)
    
    corr = np.corrcoef(y_true.flatten(), y_pred.flatten())[0, 1]
    
    ss_total = np.sum((y_true - np.mean(y_true))**2)
    ss_res = np.sum((y_true - y_pred)**2)
    r2 = 1 - (ss_res / ss_total)
    
    return mse, corr, r2
```

```python
X, y, X_test = load_data()

w = ols_with_intercept(X, y)
y_pred = predict_with_intercept(X, w)

mse, corr, r2 = compute_metrics(y, y_pred)

print("MSE:", mse)
print("Correlation:", corr)
print("R2 Score:", r2)
```
